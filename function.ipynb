```python
def fib(n):
    if n in [0,1]:
        return n
    else:
        return fib(n-1)+fib(n-2)
```

```python
fib(10)
```

```python
def fibb(n):
    a = 0
    b = 1
    for i in range(n):
        yield a,i
        a,b = b, a+b
```

```python
for i in fibb(11):
    print(i)
```

```python
def fib1(n):
    a = 0
    b = 1
    output = []
    for i in range(n):
        output.append(a)
        a,b = b, a+b
    return output    
```

```python
fib1(11)
```
