```python
'''3 . Write a fun which will take input as a dict and give me out as a list of all the values 
even in case of 2 level nesting it should work . '''
def getValuesInList(ip_val):
    l = []
    if type(ip_val) == dict:
        l.append(ip_val.values())
    for i in ip_val:
        if type(i) == dict:
            l.append(i.values())
    return l   
inp_val = {'name':'faiz','mob':'9873300865','a':{'add':'topkhana bazar','city':'Delhi'}}
getValuesInList(inp_val)    
```

```python
# 11 . write a function which can read video file and play for you . 
import cv2

def playVideo(video_path):
    video = cv2.VideoCapture(video_path)
    while (video.isOpened()):
        # capturing the video frame-by-frame
        ret, frame = video.read()
        if ret == True:
            cv2.imshow('Frame',frame)  
            if cv2.waitKey(25) & 0xFF == ord('x'):
                break
        else:
            break
    video.release()
    cv2.destroyAllWindows()
    
```

```python
playVideo('bigbuckbunny.mp4')
```

```python
#15 . write a func bywhcih i can send a mail to anyone 
import smtplib
def sendEmail(user,passwd):
    server = smtplib.SMTP('smtp.gmail.com',587)
    server.starttls()
    server.login(user,passwd)
    rx_mail = input("enter receiver mail id: ")
    msg = input("enter the mail(body)")
    server.sendmail(user,rx_mail,msg)
    server.close()
    print("task completed")
```

```python
#16. write a func to read a complete PDf file . 
import PyPDF2 as pdf
def readFile(filename):
    f = open(filename+'.pdf','rb')
    f_read = pdf.PdfFileReader(f)
    f_page = f_read.getPage(0)
    print(f_page.extractText())
    f.close()
    
inp_filename = input("enter your file name: ")    
readFile(inp_filename)    
```

```python
#18 . write a function which can help you to filter only word file from a directory . 
import os
import re
def getWordFiles(file_path):
    a = os.walk(file_path)
    l = []
    dir_lst = list(a)
    
    pattern = re.compile(r"\.doc$")
    
    for lst in dir_lst:
        if type(lst) == tuple:
            for dirc in lst:
                if type(dirc) == list:
                    for file in dirc:
                        matches = pattern.finditer(file)
                        for match in matches:
                            l.append(file)
    return l
directory_path = input("Please enter the directory path to all list  .doc files in it: ")
getWordFiles(directory_path)
#C:\\Users\\fazlu\\Working_space\\class\python
```
